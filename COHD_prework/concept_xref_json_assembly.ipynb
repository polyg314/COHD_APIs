```python
import pandas as pd
import requests as re
import json

data = pd.read_csv('concepts.txt', sep="\t")
data
```

```python
# concept_xref_json = []

for i in range(62298,data.shape[0]):
    if(i%10 == 0):
        print(i)
    concept_id = str(data["concept_id"][i])
    res = re.get('http://tr-kp-clinical.ncats.io/api/omop/xrefFromOMOP?concept_id=' + concept_id)
    if(res.status_code == 200):
        res_json = json.loads(res.text)
    xrefs = []
    
    for x in res_json['results']:
        if x['target_curie'] not in xrefs: 
            xrefs.append(x['target_curie'])
    if(len(xrefs) == 0):
        xrefs.append(str(data["vocabulary_id"][i]) + ":" + str(data["concept_code"][i]))
    xref_dict = {}
    for y in xrefs:
        y = y.split(':')
        key = y[0]
        value = y[1]
        if key in xref_dict:
            xref_dict[key] = [xref_dict[key]] if isinstance(xref_dict[key], str) else xref_dict[key]
            xref_dict[key].append(value)
        else:
            xref_dict[key] = value
    concept_id_dict = {
        "_id" : concept_id,
        "concept_name": data["concept_name"][i],
        "domain_id": data["domain_id"][i],
        "concept_class_id": data["concept_class_id"][i],
        "xrefs": xref_dict
    }
    concept_xref_json.append(concept_id_dict)
```

```python
with open('concept_xref_backup.json', 'w', encoding='utf-8') as f:
    json.dump(concept_xref_json, f, ensure_ascii=False, indent=4)
```
